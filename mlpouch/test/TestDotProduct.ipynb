```python
from mlpouch.similarity import DotProduct
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("data/movies_clean.csv")
del movies['Unnamed: 0']
movies.head()
```

```python
movies_mat = movies.iloc[:, 4:]
movies_mat.head()
```

```python
dp = DotProduct()
movies_dot = dp.fit_transform(movies_mat)
movies_dot
```
